## Pandas Practice

‚úçüèª  (10pts) From the standard penguins dataset, answer this question: what percentage of female penguins from Dream have a body mass greater than $4000$ grams? (Use only Pandas attributes and methods. Show your code as well as your answer.)

```python
import seaborn as sns
import pandas as pd

peng = sns.load_dataset("penguins")

peng1 = peng[peng["sex"] == "Female"]
peng2 = peng1[peng1["island"] == "Dream"]
peng3 = peng2[peng2["body_mass_g"] > 4000]

perc = len(peng3) / len(peng2)
perc*100
```

‚úÖAnswer:

*Thus 1.64 percent of female penguins from Dream have a body mass greater than $4000$ grams.*

## EDA: Seaborn and Correlations

‚úçüèª  (10pts) Read in the mpg dataset and perform EDA on it. Specifically:
* get the basic statistics from the df,


```python
import seaborn as sns
mpg = sns.load_dataset("mpg")
mpg.head()
```

```python
mpg.describe()
```

* make a `displot` of horsepower versus mpg, including a rug plot and with a kde,


```python

sns.displot(data = mpg, x = "mpg", y = "horsepower",kind = "kde",rug = True)
```

‚úÖAnswer:

*We can see that, the higher the horsepower, the great the acceleration.*

* make a Seaborn `heatmap` of the correlations using Pandas' `.corr`, with annotations and your choice of colormap,


```python
import pandas as pd

corre = mpg.corr()

type(corre)

sns.heatmap(mpg.corr(), annot=True, cmap="YlGnBu")
```

‚úÖAnswer:

*We can see that there are high correlation between cylinder number, displacement, horsepower and weight.*

* make a `regplot` ([or, `lmplot`](https://seaborn.pydata.org/tutorial/regression.html)) of horsepower versus mpg using the lowess smoother; choose good colors,


```python
sns.regplot(data=mpg, x="horsepower", y="mpg", lowess=True, line_kws={"color": "C5"})
```

‚úÖAnswer:

*We can see that, the relationship between horsepower and mpg is linear, and can be explained by a linear model, only a few outliers can be found on this model.*

* design a `FacetGrid` of your choice.


```python
fig = sns.FacetGrid(mpg, col="cylinders", hue="origin")
fig.map(sns.regplot, "horsepower", "mpg", lowess=True)
fig.add_legend()
```

‚úÖAnswer:

* 3 cylinder cars are only made in Japan;

* 5 cylinder cars are only made in Europe;

* 8 cylinder cars are only made by USA.

* Generally the more cylinder a car has the greater the horsepower it has.
