```python
import seaborn as sns
from scipy.stats import zscore
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
digi = pd.read_csv("digimon.csv")
```

```python
digi
```

```python
pd.plotting.parallel_coordinates(digi[["Type", "Lv50 Atk", "Lv50 Spd", "Lv50 Int", "Lv50 Def"]],"Type")
```

```python
pd.plotting.parallel_coordinates(digi[["Stage", "Lv50 Atk", "Lv50 Spd", "Lv50 Int", "Lv50 Def"]],"Stage")
```

```python
pd.plotting.parallel_coordinates(digi[["Attribute", "Lv50 Atk", "Lv50 Spd", "Lv50 Int", "Lv50 Def"]],"Attribute")
```

âœ…Conclusion: The numbers don;t depend on "attribute" or "type" but depend on "stage" more.

```python

numeric_cols = digi.select_dtypes(include=[np.number]).columns

digi_s = digi[numeric_cols].apply(zscore)

digi_s["Stage"] = digi["Stage"]
pd.plotting.parallel_coordinates(digi_s, "Stage")
```

```python
numeric_cols = digi.select_dtypes(include=[np.number]).columns

digi_s2 = digi[numeric_cols].apply(zscore)

digi_s2["Type"] = digi["Type"]
pd.plotting.parallel_coordinates(digi_s2, "Type")
```
